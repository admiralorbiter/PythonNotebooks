# Boolean Operators ##

| Symbols | Meaning | Example |
|----|----|----|
| < | less than | 3 < 5 |
| > | greater than | age > 30 |
| <= | less than or equal | 30.0 <= 30 |
| >= | greater than or equal | age >= 30 |
| == | Equality comparison | name == "Kendall" |
| != | Not equal | class != "CS201" |

** Note that == is equality, = is assignment **

````python
height = 10

print(height > 9)
````

````python
print(height < 7)
````

````python
print(height >= 5.7)
````

````python
print(height == 8)
````

````python
print(height != 7)
````

# If Else

Indentation has meaning in python.  Each item in a suite must be indended the same.

````python
test_score = int(input("What did you get on the test?"))

if test_score >= 90:
    print("You got an A")
else:
    print("Work some more ")

print("Rest of program")
````

## elif

```
if <condition>:
    suite1
elif <condition>:
    suite2
elif <condition>:
    suite3
else:
    suite4
    
```

````python
test_score = int(input("What did you get on the test?"))

if test_score >= 80:
    print("You got an b")
elif test_score >= 90:
    print("You got a A")
elif test_score >= 70:
    print("You got a C")
elif test_score >= 60:
    print("You got a D")
else:
    print("Work Harder")

print("rest of program")
````

# And Operator

| op1 | op2 | ```op1 and op2``` |
|----|----|----|
| False | False | False|
| False | True | False |
| True | False | False |
| True | True | True |



````python
age = 8
day_of_week = 'Saturday'

if age >= 21 and day_of_week== 'Saturday':
    print("Go see a movie")
````

# Or Operator

| op1 | op2 | ```op1 or op2``` |
|----|----|----|
| False | False | False|
| False | True | True |
| True | False | True |
| True | True | True |

````python
age = 17
income = 10000

if age > 18 or income > 50000:
    print("Go visit hawaii")
    
````

# Not Operator

| op1 | not op1 |
|----|----|
| True | False |
| False | True |

````python
is_minor = True

print("Person is an adult", not is_minor)
````
