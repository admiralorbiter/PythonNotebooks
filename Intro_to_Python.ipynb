# Data types

Programming languages handle different data types.  Each has it's own specific uses which model our own thinking.

| Data type | Description |
|-----------|-------------|
| str string | A literal string representing text.  Can begin and end with " or ' |
| int - integer | A numeric value with no decimal portion |
| float | A floating point value.  A number with decimal precision |
| bool ( Boolean ) | Booleans are True False.  They only have 2 states and represent basic logic |
| list | Lists are collections of objects |
| dict - dictionary | A dictionary has a key value pair of collection of objects |


## Integers

Integers are a numeric value and do not contain any decimal precision.

```python
speed = 60     # velocity in miles per hour
distance = 120 # how far we need to travel for our trip

time = distance / speed  # How long it will take us
print("You will travel", distance, "in", time, "hours with the current rate of ", speed, "mph")
```

# Operators

| Operator | It's function |
|----------|---------------|
| + | Addition  ( a + b ) |
| - | Subtraction ( a - b ) |
| * | Multiplication ( a * b ) |
| / | Division ( a / b ) |
| ** | Exponent a \*\* b.    2\*\*3 is 2\*2\*2 evaluates to 8 |
| // | Division with only the quotient remainder   5 // 2  would be just 2 |
| % | Modulus operator.  Gives the remainder  12 % 10 evalutes to 2 |


```python
a = 5
b = 12
c = a + b
print(c)
```

```python
a = 5
b = 3

c = a - b
print(c)
```

```python
a = 10
b = 3

value = a / b
quotient = a // b
remainder = a % b
print(value)
print(quotient)
print(remainder)
```

```python
a = 10

b = 10 ** 3

print(b)
```
